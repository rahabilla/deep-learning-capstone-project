```python
import os
import math
import json
import librosa


SAMPLE_RATE = 22050
SAMPLES_PER_TRACK = SAMPLE_RATE * 30  # 30 seconds per track

def save_mfcc(dataset_path, json_path, num_mfcc=13, n_fft=2048, hop_length=512, num_segments=10):
    data = {
        "mapping": [],
        "labels": [],
        "mfcc": []
    }

    samples_per_segment = int(SAMPLES_PER_TRACK / num_segments)
    expected_num_mfcc_vectors = math.ceil(samples_per_segment / hop_length)

    for i, (dirpath, dirnames, filenames) in enumerate(os.walk(dataset_path)):
        if dirpath != dataset_path:
            label = os.path.basename(dirpath)
            data["mapping"].append(label)
            print(f"\nProcessing: {label}")

            for f in filenames:
                file_path = os.path.join(dirpath, f)

                if not file_path.lower().endswith((".wav", ".mp3", ".au")):
                    continue

                try:
                    signal, sr = librosa.load(file_path, sr=SAMPLE_RATE)

                    for d in range(num_segments):
                        start = samples_per_segment * d
                        finish = start + samples_per_segment

                        mfcc = librosa.feature.mfcc(
                            y=signal[start:finish],
                            sr=sr,
                            n_mfcc=num_mfcc,
                            n_fft=n_fft,
                            hop_length=hop_length
                        ).T

                        if len(mfcc) == expected_num_mfcc_vectors:
                            data["mfcc"].append(mfcc.tolist())
                            data["labels"].append(i - 1)
                            print(f"{file_path}, segment: {d + 1}")
                except Exception as e:
                    print(f"❌ Error processing {file_path}: {e}")

    with open(json_path, "w") as fp:
        json.dump(data, fp, indent=4)


DATASET_PATH = r"C:\Users\harip\Downloads\archive (2)\Data\genres_original"
JSON_PATH = r"C:\Users\harip\Downloads\mfcc_data.json"


save_mfcc(DATASET_PATH, JSON_PATH, num_segments=10)
```

```python
DATA_PATH = r"C:\Users\harip\Downloads\mfcc_data.json"
```

```python
def load_data(data_path):

    with open(data_path, "r") as f:
        data = json.load(f)

    # convert lists to numpy arrays
    X = np.array(data["mfcc"])
    y = np.array(data["labels"])

    print("Data succesfully loaded!")

    return  X, y
```

```python
# load data
X, y = load_data(DATA_PATH)
```

```python
X.shape
```

```python
# create train, validation and test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)

# add an axis to input sets
X_train = X_train[..., np.newaxis]
X_test = X_test[..., np.newaxis]
```

```python
X_train.shape
```

```python
input_shape = (X_train.shape[1], X_train.shape[2], 1)
```

```python
from keras.regularizers import l2
# build the CNN
model_cnn = keras.Sequential()

# 1st conv layer
model_cnn.add(keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=input_shape))
model_cnn.add(keras.layers.MaxPooling2D((3, 3), strides=(2, 2), padding='same'))
model_cnn.add(keras.layers.BatchNormalization())


# 2nd conv layer
model_cnn.add(keras.layers.Conv2D(64, (3, 3), activation='relu',kernel_initializer='he_uniform',kernel_regularizer=l2(0.0001)))
model_cnn.add(keras.layers.MaxPooling2D((3, 3), strides=(2, 2), padding='same'))
model_cnn.add(keras.layers.BatchNormalization())
model_cnn.add(keras.layers.Dropout(0.3))

# 3rd conv layer
model_cnn.add(keras.layers.Conv2D(128, (2, 2), activation='relu',kernel_initializer='he_uniform',kernel_regularizer=l2(0.0001)))
model_cnn.add(keras.layers.MaxPooling2D((2, 2), strides=(2, 2), padding='same'))
model_cnn.add(keras.layers.BatchNormalization())
model_cnn.add(keras.layers.Dropout(0.3))

# flatten output and feed it into dense layer
model_cnn.add(keras.layers.Flatten())
model_cnn.add(keras.layers.Dense(128, activation='relu',kernel_initializer='he_uniform',kernel_regularizer=l2(0.0001)))
model_cnn.add(keras.layers.BatchNormalization())
model_cnn.add(keras.layers.Dropout(0.3))

# output layer
model_cnn.add(keras.layers.Dense(10, activation='softmax'))
```

```python
# compile model
optimiser = keras.optimizers.Adam(learning_rate=0.0001)
model_cnn.compile(optimizer=optimiser,
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
from tensorflow.keras.callbacks import EarlyStopping
callbacks = [
             EarlyStopping(patience=3)
]
```

```python
model_cnn.summary()
```

```python
# train model
history = model_cnn.fit(X_train, y_train, validation_data=(X_test, y_test), batch_size=32, epochs=100,callbacks=callbacks)
```

```python
import matplotlib.pyplot as plt

def plot_history(history):
    # Create a figure with 2 subplots and increased size
    fig, axs = plt.subplots(2, figsize=(10, 8))

    # Accuracy subplot
    axs[0].plot(history.history["accuracy"], label="Train Accuracy")
    axs[0].plot(history.history["val_accuracy"], label="Validation Accuracy")
    axs[0].set_ylabel("Accuracy")
    axs[0].set_title("Accuracy over Epochs")
    axs[0].legend(loc="lower right")

    # Loss subplot
    axs[1].plot(history.history["loss"], label="Train Loss")
    axs[1].plot(history.history["val_loss"], label="Validation Loss")
    axs[1].set_ylabel("Loss")
    axs[1].set_xlabel("Epoch")
    axs[1].set_title("Loss over Epochs")
    axs[1].legend(loc="upper right")

    # Fix layout
    plt.tight_layout()
    plt.show()

# Call the function
plot_history(history)
```

```python
def process_input(audio_file, track_duration):
    import math, librosa, numpy as np

    SAMPLE_RATE = 22050
    NUM_MFCC = 13
    N_FTT = 2048
    HOP_LENGTH = 512
    SAMPLES_PER_TRACK = SAMPLE_RATE * track_duration
    NUM_SEGMENTS = 10

    samples_per_segment = int(SAMPLES_PER_TRACK / NUM_SEGMENTS)

    signal, _ = librosa.load(audio_file, sr=SAMPLE_RATE)

    
    start = 0
    finish = samples_per_segment

    mfcc = librosa.feature.mfcc(
        y=signal[start:finish],
        sr=SAMPLE_RATE,
        n_mfcc=NUM_MFCC,
        n_fft=N_FTT,
        hop_length=HOP_LENGTH
    ).T

    return mfcc

# Load and reshape input
new_input_mfcc = process_input(r"C:\Users\harip\Downloads\archive (2)\Data\genres_original\blues\blues.00000.wav", 30)
X_to_predict = new_input_mfcc[np.newaxis, ..., np.newaxis]  # (1, 129, 13, 1)

# Predict
prediction = model_cnn.predict(X_to_predict)
predicted_index = np.argmax(prediction, axis=1)

# Decode
genre_dict = {
    0: "disco", 1: "pop", 2: "classical", 3: "metal", 4: "rock",
    5: "blues", 6: "hiphop", 7: "reggae", 8: "country", 9: "jazz"
}
print("Predicted Genre:", genre_dict[int(predicted_index)])
```
